# 4: Break Point

What is the break point for a 3D perceptron model?

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([2, 2, 3], [1, 2, 0], [1, 3, 3], color="blue")
ax.scatter([1, 4], [-1, 4], [2, 2], color="red")
plt.show()
```

# 7

Number of dichotomies for 2-intervals hypothesis set.

N=3: 2^3 = 8

```python
from scipy.special import comb as choose

def alt_a(N):
    return choose(N+1, 4)

def alt_b(N):
    return choose(N+1, 2) + 1

def alt_c(N):
    return choose(N+1, 4) + choose(N+1, 2) + 1

def alt_d(N):
    return choose(N+1, 4) + choose(N+1, 3) + choose(N+1, 2) + choose(N+1, 1) + 1
```

```python
print("\nA")
for i in range(1, 7):
    print(i, alt_a(i))

print("\nB")    
for i in range(1, 7):
    print(i, alt_b(i))

print("\nC")    
for i in range(1, 7):
    print(i, alt_c(i))
    
print("\nD")    
for i in range(1, 7):
    print(i, alt_d(i))
    
    
print("\nComputed by hand")
print("1 2.0")
print("2 4.0")
print("3 8.0")
print("4 16.0")
print("5 31.0")
print("6 57.0 (64 - 7)")
```

So the only alternatives left are C and "none of the above".

Reasoning for C:

Number of ways to choose two non-overlapping intervals 

plus

Number of ways to choose one interval

plus 1 for nothing being selected.
